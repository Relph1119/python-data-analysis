### 步骤9 被下单数最多商品(item)是什么?

### 步骤13 将item_price转换为浮点数

#### 步骤13.a 检查item_price的类型

```python
chipo.item_price.dtype
```

#### 步骤13.b 创建一个lambda方法，用于转换item_price的类型

```python
dollarizer = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollarizer)
```

#### 步骤13.c 检查转换后的item_price类型

```python
chipo.item_price.dtype
```

### 步骤14 在该数据集对应的时期内，收入(revenue)是多少？

```python
chipo['sub_total'] = round(chipo['item_price'] * chipo['quantity'],2)
revenue = chipo['sub_total'].sum()
print('Revenue was: $' + str(revenue))
```

### 步骤16 每一单(order)对应的平均总价是多少？
