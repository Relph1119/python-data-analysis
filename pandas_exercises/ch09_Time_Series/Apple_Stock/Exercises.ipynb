# 探索Apple公司股价数据

## 步骤1 导入必要的库

```python
import pandas as pd
import numpy as np

# visualization
import matplotlib.pyplot as plt

%matplotlib inline
```

## 从如下地址导入数据集

```python
path = '../../exercise_data/Apple_stock.csv'
```

## 步骤3 读取数据并保存为一个名叫apple的数据框

字段名称 | 解释说明
--- | ---
Date | 日期
Open | 开盘价位
High | 当日最高
Low | 当日最低
Close | 收盘价位
Volume | 成交量
Adj Close | 经过调整的收盘价

```python
apple = pd.read_csv(path)
apple.head()
```

## 步骤4 查看每一列的数据类型

```python
apple.dtypes
```

## 步骤5 将Date这列转换为datetime类型

```python
apple.Date = pd.to_datetime(apple.Date)
apple['Date'].head()
```

## 将Date设置为索引

```python
apple = apple.set_index('Date')
apple.head()
```

## 步骤7 有重复的日期吗？

```python
apple.index.is_unique
```

## 步骤8 索引列是按照最近的日期排列的，将其设置为升序（按照最远的日期排列）

```python
apple.sort_index(ascending=True).head()
```

## 步骤9 找到每个月的最后一个交易日(business day)

```python
apple_month = apple.resample('BM').mean()
apple_month.head()
```

## 步骤10 数据集中最早的日期和最晚的日期相差多少天？

```python
(apple.index.max() - apple.index.min()).days
```

## 步骤11 在数据中一共有多少个月？

```python
apple_months = apple.resample('BM').mean()
len(apple_months.index)
```

## 步骤12 按照时间顺序可视化Adj Close值，设置图形大小为13.5 x 9 inches

```python
apple_open = apple['Adj Close'].plot(title='Apple Stock')

fig = apple_open.get_figure()
fig.set_size_inches(13.5, 9)
```
