# 练习-数据分组

## 探索酒类消费数据

### 步骤1 导入必要的库

```python
import pandas as pd
```

### 步骤2 从以下地址导入数据

```python
path ='../../exercise_data/drinks.csv'
```

### 步骤3 将数据框命名为drinks

字段名称 | 解释说明
--- | ---
country	| 国家
beer_servings |	啤酒消费
spirit_servings | 烈酒消费
wine_servings | 红酒消费
total_litres_of_pure_alchol | 消费的纯酒精统计
continent | 大陆

```python
drinks = pd.read_csv(path)
drinks.head()
```

### 步骤4 哪个大陆(continent)平均消耗的啤酒(beer)更多？

```python
drinks.groupby('continent').beer_servings.mean()
```

### 步骤5 打印出每个大陆(continent)的红酒消耗(wine_servings)的描述统计值 

```python
drinks.groupby('continent').wine_servings.describe()
```

### 步骤6 打印出每个大陆(continent)每种酒类别的消耗平均值 

```python
drinks.groupby('continent').mean()
```

### 步骤7 打印出每个大陆(continent)每种酒类别的消耗中位数

```python
drinks.groupby('continent').median()
```

### 步骤8 打印出每个大陆对spirit饮品(spirit_servings)消耗的平均值，最大值和最小值

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'max', 'min'])
```
