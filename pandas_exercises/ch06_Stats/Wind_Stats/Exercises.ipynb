# 探索风速数据(Wind Statistics)

## 数据说明
在`wind.data`文件中的数据格式如下：

```python
"""
Yr Mo Dy   RPT   VAL   ROS   KIL   SHA   BIR   DUB   CLA   MUL   CLO   BEL   MAL
61  1  1 15.04 14.96 13.17  9.29   NaN  9.87 13.67 10.25 10.83 12.58 18.50 15.04
61  1  2 14.71   NaN 10.83  6.50 12.62  7.67 11.50 10.04  9.79  9.67 17.54 13.83
61  1  3 18.50 16.88 12.33 10.13 11.17  6.17 11.25   NaN  8.50  7.67 12.75 12.71
"""
```

前三列是年、月和日，剩下的12列表示当天在爱尔兰12个不同地点的平均风速。

## 步骤1 导入必要的库

```python
import pandas as pd
```

## 步骤2 从以下地址导入数据

```python
path = "../../exercise_data/wind.data"
```

## 步骤3 将数据保存在`data`的数据框中，并设置前三列为datetime类型的索引

```python
data = pd.read_table(path, sep = '\s+', parse_dates=[[0, 1, 2]])
data.head()
```

## 步骤4 2061年？我们真的有这一年的数据？创建一个函数并用它去修复这个bug

```python
import datetime

def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

data.Yr_Mo_Dy = data.Yr_Mo_Dy.apply(fix_century)
data.head()
```

## 步骤5 将日期设为索引，注意数据类型，应该是datetime64[ns]

```python
data.Yr_Mo_Dy = pd.to_datetime(data.Yr_Mo_Dy)

data = data.set_index(data.Yr_Mo_Dy.name)

data.head()
```

## 步骤6 对应每一个location，一共有多少数据值缺失
**在下面的计算中，应该将这些缺失值删掉**

```python
data.isnull().sum()
```

## 步骤7 对应每一个location，一共有多少完整的数据值

```python
data.notnull().sum()
```

## 步骤8 计算所有地点所有时间内风速的平均值

```python
data.mean().mean()
```

## 步骤9 创建一个名为loc_stats的数据框，计算每个location所有时间内的风速最小值(min)，最大值(max)，平均值(mean)和标准差(std)

```python
loc_stats = pd.DataFrame()

loc_stats['min'] = data.min()
loc_stats['max'] = data.max()
loc_stats['mean'] = data.mean()
loc_stats['std'] = data.std()

loc_stats
```

## 步骤10 创建一个名为day_stats的数据框，计算所有location每一天的风速最小值(min)，最大值(max)，平均值(mean)和标准差(std)

```python
day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis=1)
day_stats['max'] = data.max(axis=1)
day_stats['mean'] = data.mean(axis=1)
day_stats['std'] = data.std(axis=1)

day_stats.head()
```

## 步骤11 对于每一个location，计算一月份的平均风速
**注意：1961年的1月和1962年的1月应该区别对待**

```python
data[data.index.month == 1].mean()
```

## 步骤12 对于数据记录按照年为频率取样

```python
data.groupby(data.index.to_period('A')).mean()
```

## 步骤13 对于数据记录按照月为频率取样

```python
data.groupby(data.index.to_period('M')).mean()
```

## 步骤14 对于数据记录按照周为频率取样

```python
data.groupby(data.index.to_period('W')).mean()
```

## 步骤15 对于数据记录按照周为频率，计算前52周风速的最小值(min)、最大值(max)、平均值(mean)和标准差(std)（假设从1961年1月2日开始为第一周）

```python
weekly = data.resample('W').agg(['min', 'max', 'mean', 'std'])

weekly.loc[weekly.index[1:53], 'RPT':'MAL'].head(10)
```
