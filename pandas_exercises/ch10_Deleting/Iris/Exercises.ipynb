# 探索Iris纸鸢花数据

## 步骤1 导入必要的库

```python
import pandas as pd
import numpy as np
```

## 步骤2 从如下地址导入数据集

```python
path ='../../exercise_data/iris.csv'
```

## 步骤3 将数据集保存为iris数据框

```python
iris = pd.read_csv(path)
iris.head()
```

## 步骤4 创建数据框的列名称

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class

iris = pd.read_csv(path,names = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'])
iris.head()
```

## 步骤5 数据框中有缺失值吗？

```python
iris.isnull().sum()
```

## 步骤6 将列petal_length的第10到29行设置为缺失值(NaN)

```python
iris.iloc[10:29, 2:3] = np.nan
iris.head(20)
```

## 步骤7 将缺失值全部替换为1.0

```python
iris.petal_length.fillna(1, inplace=True)
iris.head(20)
```

## 步骤8 删除列class

```python
del iris['class']
iris.head()
```

## 步骤9 将数据框前三行设置为缺失值

```python
iris.iloc[0:3, :] = np.nan
iris.head()
```

## 步骤10 删除有缺失值(NaN)的行

```python
iris = iris.dropna(how='any')
iris.head()
```

## 步骤11 重新设置索引

```python
iris = iris.reset_index(drop=True)
iris.head()
```
